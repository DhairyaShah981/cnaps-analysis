```python
# convert the dataset folder of images in gz format

import tarfile
import gzip
import os

def create_gzipped_tarball(source_dir, output_filename):
    # Create a tarball and compress it in gzip format
    with tarfile.open(output_filename + '.tar', 'w') as tar:
        # Add directory to tarball
        tar.add(source_dir, arcname=os.path.basename(source_dir))

    # Compress the tarball
    with open(output_filename + '.tar', 'rb') as f_in:
        with gzip.open(output_filename + '.tar.gz', 'wb') as f_out:
            f_out.writelines(f_in)

    # Optionally, remove the uncompressed tarball
    os.remove(output_filename + '.tar')

# Usage
source_directory = 'dataset'  # Replace with your folder path
output_file = 'celeb_archive_images'  # Output file name without extension
create_gzipped_tarball(source_directory, output_file)
```

```python
class ExtraDatasetConverter(DatasetConverter):
    def create_splits(self):
        class_names = sorted(os.listdir(self.data_root))
        return {'train': [], 'valid': [], 'test': class_names}

    def create_dataset_specification_and_records(self):
        splits = self.get_splits(force_create=True)
        self.classes_per_split[learning_spec.Split.TRAIN] = len(splits['train'])
        self.classes_per_split[learning_spec.Split.VALID] = len(splits['valid'])
        self.classes_per_split[learning_spec.Split.TEST] = len(splits['test'])

        for class_id, class_name in enumerate(splits['test']):
            print('Creating record for class ID {} ({})'.format(class_id, class_name))
            class_directory = os.path.join(self.data_root, class_name)
            class_records_path = os.path.join(self.records_path, self.dataset_spec.file_pattern.format(class_id))
            self.class_names[class_id] = class_name
            self.images_per_class[class_id] = write_tfrecord_from_directory(class_directory, class_id, class_records_path)
```

```python
datasrc_path = '/media/cvig/zoneb/pml/cnaps/meta-dataset/data'
records_path = '/media/cvig/zoneb/pml/cnaps/meta-dataset/meta_dataset/records'
splits_path = '/media/cvig/zoneb/pml/cnaps/meta-dataset/meta_dataset/dataset_conversion/splits'

converter = ExtraDatasetConverter(
        name='celebcars',
        data_root=os.path.join(datasrc_path, 'celebcars'),
        has_superclasses=False,
        records_path=os.path.join(records_path, 'celebcars'),
        split_file = os.path.join(splits_path, 'celebcars.json'),
        random_seed=22
    )

converter.convert_dataset()
```
