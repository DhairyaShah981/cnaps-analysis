```python
from tueplots import bundles
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams.update(bundles.beamer_moml())

# Also add despine to the bundle using rcParams
plt.rcParams["axes.spines.right"] = False
plt.rcParams["axes.spines.top"] = False

# Increase font size to match Beamer template
plt.rcParams["font.size"] = 16
# Make background transparent
plt.rcParams["figure.facecolor"] = "none"
# use font size Roboto Condensed
plt.rcParams["font.family"] = "sans-serif"


# For saving the plot:
# plt.savefig('filename.pdf', bbox_inches='tight', transparent=True)
```

```python

def plot_tue(filename, training_steps, train_losses, train_accuracies, validation_steps, val_accuracies, fully_trained_testing_accuracy, best_validation_testing_accuracy):
        # Set the style for seaborn
    sns.set(style="whitegrid")

    # Create subplots
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

    # Plot 1: Training Accuracy
    axes[0, 0].plot(training_steps, train_accuracies, marker='o', label='Training Accuracy')
    axes[0, 0].set_title('Training Accuracy vs Epochs')
    axes[0, 0].set_xlabel('Epochs')
    axes[0, 0].set_ylabel('Accuracy')
    axes[0, 0].legend()

    # Plot 2: Training Loss
    axes[0, 1].plot(training_steps, train_losses, marker='o', color='r', label='Training Loss')
    axes[0, 1].set_title('Training Loss vs Epochs')
    axes[0, 1].set_xlabel('Epochs')
    axes[0, 1].set_ylabel('Loss')
    axes[0, 1].legend()

    # Plot 3: Validation Accuracy
    axes[1, 0].plot(validation_steps, val_accuracies, marker='o', label='Validation Accuracy')
    axes[1, 0].set_title('Validation Accuracy vs Epochs')
    axes[1, 0].set_xlabel('Epochs')
    axes[1, 0].set_ylabel('Accuracy')
    axes[1, 0].legend()

    # Plot 4: Testing Accuracy
    axes[1, 1].axhline(y=fully_trained_testing_accuracy, linestyle='--', color='b', label='Fully Trained Testing Accuracy')
    axes[1, 1].axhline(y=best_validation_testing_accuracy, linestyle='--', color='g', label='Best Validation Testing Accuracy')
    axes[1, 1].set_title('Testing Accuracy')
    axes[1, 1].set_ylabel('Accuracy')
    axes[1, 1].legend()

    # Adjust layout to prevent clipping of labels
    plt.tight_layout()
    plt.savefig(filename, bbox_inches='tight', transparent=True)
```

```python
# make a plot for comparision across different shots in celeb10
# using subplots, it should have only two plots, one for training accuracy and one for testing accuracy on fully trained model

def plot_shots(filename, training_steps, train_accuracies, fully_trained_testing_accuracies):
        # Set the style for seaborn
    sns.set(style="whitegrid")

    # Create subplots
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 5))

    # Plot 1: Training Accuracy
    axes[0].plot(training_steps, train_accuracies[0], marker='o', label='Context Size = 1')
    axes[0].plot(training_steps, train_accuracies[1], marker='o', label='Context Size = 3')
    axes[0].plot(training_steps, train_accuracies[2], marker='o', label='Context Size = 5')
    axes[0].set_title('Training Accuracy vs Epochs')
    axes[0].set_xlabel('Epochs')
    axes[0].set_ylabel('Accuracy')
    axes[0].legend()

    # Plot 4: Testing Accuracy
    axes[1].axhline(y=fully_trained_testing_accuracies[0], linestyle='--', color='b', label='Context Size = 1')
    axes[1].axhline(y=fully_trained_testing_accuracies[1], linestyle='--', color='g', label='Context Size = 3')
    axes[1].axhline(y=fully_trained_testing_accuracies[2], linestyle='--', color='r', label='Context Size = 5')
    axes[1].set_title('Testing Accuracy')
    axes[1].set_ylabel('Accuracy')
    axes[1].legend()

    # Adjust layout to prevent clipping of labels
    plt.tight_layout()
    plt.savefig(filename, bbox_inches='tight', transparent=True)
```

```python
# make a plot for comparision across different shots in celeb10
# using subplots, it should have only two plots, one for training accuracy and one for testing accuracy on fully trained model

def plot_datasets(filename, training_steps, train_accuracies, fully_trained_testing_accuracies):
        # Set the style for seaborn
    sns.set(style="whitegrid")

    # Create subplots
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 5))

    # Plot 1: Training Accuracy
    axes[0].plot(training_steps, train_accuracies[0], marker='o', label='Celeb10')
    axes[0].plot(training_steps, train_accuracies[1], marker='o', label='Celeb150')
    axes[0].plot(training_steps, train_accuracies[2], marker='o', label='CelebCars')
    axes[0].set_title('Training Accuracy vs Epochs')
    axes[0].set_xlabel('Epochs')
    axes[0].set_ylabel('Accuracy')
    axes[0].legend()

    # Plot 4: Testing Accuracy
    axes[1].axhline(y=fully_trained_testing_accuracies[0], linestyle='--', color='b', label='Celeb10')
    axes[1].axhline(y=fully_trained_testing_accuracies[1], linestyle='--', color='g', label='Celeb150')
    axes[1].axhline(y=fully_trained_testing_accuracies[2], linestyle='--', color='r', label='CelebCars')
    axes[1].set_title('Testing Accuracy')
    axes[1].set_ylabel('Accuracy')
    axes[1].legend()

    # Adjust layout to prevent clipping of labels
    plt.tight_layout()
    plt.savefig(filename, bbox_inches='tight', transparent=True)
```
